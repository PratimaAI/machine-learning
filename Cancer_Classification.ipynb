# Train Test split

```python
X= df.drop('benign_0__mal_1',axis=1)
y= df['benign_0__mal_1']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler= MinMaxScaler()
```

```python
X_train= Scaler.fit_transform(X_train)
X_test= Scaler.transform(X_test)
```

# Creating the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model= Sequential()

model.add(Dense (30, activation= 'relu'))
model.add(Dense (15, activation= 'relu'))

#Binary classificayion problem 

model.add(Dense (1, activation= 'sigmoid'))

model.compile(optimizer= 'adam', loss='binary_crossentropy')
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss
```

```python
loss.plot()
```

```python
#Because of too many epoch(=600), validation loss is increasing. This is the issue of overfitting.
```

```python
# Dropout method and Earlystopping will be used
```

# Earlystopping

```python
model= Sequential()

model.add(Dense (30, activation= 'relu'))
model.add(Dense (15, activation= 'relu'))

#Binary classificayion problem 

model.add(Dense (1, activation= 'sigmoid'))

model.compile(optimizer= 'adam', loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop= EarlyStopping(monitor='val_loss', mode= 'min', verbose=1,patience=25)
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test), callbacks=early_stop)
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss
```

```python
model_loss.plot()
```

# Dropout

```python
model= Sequential()

model.add(Dense (30, activation= 'relu'))
model.add(Dropout (0.5))

model.add(Dense (15, activation= 'relu'))
model.add(Dropout (0.5))

#Binary classificayion problem 

model.add(Dense (1, activation= 'sigmoid'))

model.compile(optimizer= 'adam', loss='binary_crossentropy')
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test), callbacks=early_stop)
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predictions= model.predict(X_test)
```

```python
predictions = (predictions > 0.5).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
